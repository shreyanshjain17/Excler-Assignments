# Simple Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
exp = np.array([2,4,3,9,12,14,10,2,6,6])
salary = np.array([50,70,55,75,120,150,75,40,80,70])
```

```python
plt.scatter(exp,salary)
plt.xlabel("Experiance")
plt.ylabel("Salary")
plt.title("Relation between Experiance & Salary")
plt.show()
```

# 



```python
x = exp
y = salary
```

```python
b1 = (sum((x-x.mean())*(y-y.mean())))/(sum((x-x.mean())**2))
b0 = y.mean()-b1*x.mean()
```

```python
b0,b1
```

```python
y_hat = b0+b1*x
```

```python
pd.DataFrame({"Actual":y,"Predicted":y_hat,"Residual":y-y_hat})
```

# Meas Square Error

```python
mse = (sum((y-y_hat)**2))/len(y)
mse
```

# R Squared

```python
r2 = 1 - ((sum((y-y_hat)**2))/(sum((y-y.mean())**2)))
r2
```

```python
print("Finally we have accuracy is : ",r2*100)
```

# Linear Regression with Sciket-learn library

Sklearn: cleaning, processing, modeling

```python
X = x.reshape(-1,1)
```

```python
X.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X)
```

```python
pd.DataFrame({"Actual":y,"Predicted":y_pred,"Residual":y-y_hat})
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y,y_pred)
mse
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y,y_pred)
r2
```
