```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

exp = np.array([2,4,3,9,12,14,10,2,6,6])
salary = np.array([50,70,55,75,120,150,75,40,80,70])

x= exp
y= salary
W= x.reshape(-1,1)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(W,y)

y_pred= model.predict(W)
pd.DataFrame({"Actual":y,"Prediction":y_pred,"Residual":y-y_pred})
from sklearn.metrics import mean_squared_error
mse=  mean_squared_error(y,y_pred)

from sklearn.metrics import r2_score
r2= r2_score(y,y_pred)
print(r2)
 
plt.scatter(W,y,color="red")
plt.scatter(W,y_pred,color="green")
plt.plot(W,y_pred,color="blue")           
plt.xlabel("Salary")
plt.ylabel("Exp")
plt.title("Relationship between Salary and Exp ")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df= pd.read_csv("delivery_time.csv")
x= np.array(df["Delivery Time"])
y= df["Sorting Time"]
Q= x.reshape(-1,1)
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(Q,y)
y_pred= model.predict(Q)
pd.DataFrame({"Actual":y,"Prediction":y_pred,"Resudial":y-y_pred})

from sklearn.metrics import mean_squared_error
mse= mean_squared_error(y,y_pred)

from sklearn.metrics import r2_score
r2= r2_score(y,y_pred)
plt.scatter(Q,y,label="Actual point",color="red")
plt.scatter(Q,y_pred,label="predection point",color="blue")
plt.plot(Q,y_pred,label="Predection",color="green")
plt.xlabel("Delivery Time")
plt.ylabel("Sorting Time")
plt.title("Relationship between Delivery Time and Sorting Time")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("Salary_Data.csv")
x= np.array(df["YearsExperience"])
y= np.array(df["Salary"])

A= x.reshape(-1,1)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(A,y)
y_pred = model.predict(A)
pd.DataFrame({"Actual":y,"Predicted":y_pred,"Residual":y-y_pred})
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y,y_pred)
mse
from sklearn.metrics import r2_score
r2 = r2_score(y,y_pred)
plt.scatter(A,y,label="Actual point",color="green")
plt.scatter(A,y_pred,label="Prediction",color="red")
plt.plot(A,y_pred,label="Prediction",color="red")

plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.title("Relation between YearsExperience & Salary")
plt.legend()
plt.show()
```
