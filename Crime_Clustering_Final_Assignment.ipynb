# Hierarchical

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(df.iloc[:,1:])
df_norm
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='average'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean')
```

```python
# save clusters for chart
y_hc = hc.fit_predict(df_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
Clusters
```

```python

df_norm['h_clusterid'] = Clusters
df_norm
```

# K-mean clustering

```python
df
```

```python
X= df.iloc[:,1:]
X
```

```python
# Normalization function 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(X)
x
```

```python
from sklearn.cluster import KMeans
kmean = KMeans(n_clusters=3)
y= kmean.fit_predict(x)
y
```

```python
data = X
data["Cluster"] = y
data["Cluster"].value_counts()
```

```python
kmean.cluster_centers_   #Within a cluster sum 0f square
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[y==0,0],x[y==0,1],label="cluster-0",color="red")
plt.scatter(x[y==1,0],x[y==1,1],label="cluster-1",color="green")
plt.scatter(x[y==2,0],x[y==2,1],label="cluster-3",color="orange")
plt.scatter(kmean.cluster_centers_[:,0],kmean.cluster_centers_[:,1],label="Centroid",color="black",s=80)
plt.legend()
plt.xlabel("")
plt.ylabel("")
plt.show()
```

# Elbow

```python
wcss = []
for i in range(1,11):
    kmean = KMeans(n_clusters=i)
    kmean.fit(x)
    r = kmean.inertia_
    wcss.append(r)
```

```python
plt.plot(range(1,11),wcss)
plt.grid()
plt.xlabel("No of cluster")
plt.ylabel("Within a cluster sum of square")
plt.title("Elbow method")
plt.show()
```

```python
#Build Cluster algorithm
model = KMeans(n_clusters=4)
y = model.fit_predict(x)
dataset = X
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler().fit(df)
X = sc.transform(df)
X
```
